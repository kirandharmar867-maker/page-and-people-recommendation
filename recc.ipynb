```python
import json
def load_data(file_name):
    with open(file_name, "r")as f:
        data=json.load(f)
    return data
data=load_data("messive_data.json")
```

```python
def rec(user_id, data):
    user_friends={}
    for user in data["users"]:
        user_friends[user["id"]]=set(user["friends"])
    if user_id not in user_friends:
        return "invalid user id"

    derect_friends=user_friends[user_id]

    suggestions={}
    for friend in derect_friends:
        for mutual in user_friends[friend]:
            if mutual!=user_id and mutual not in derect_friends:
                suggestions[mutual]=suggestions.get(mutual, 0) +1
    sorted_suggestions=sorted(suggestions.items(), key=lambda x: x[1], reverse=True)
    
    return [user_id for user_id, _ in sorted_suggestions]

def page_recc(user_id, data):
    user_pages={}
    for user in data["users"]:
        user_pages[user["id"]]=set(user["liked_pages"])
    
    if user_id not in user_pages:
        return "invalid user"

    # input use liked pages
    liked_pages=user_pages[user_id]

    page_suggestion={}
    for other_user, pages in user_pages.items():
        if other_user!=user_id:
            shared_pages=liked_pages.intersection(pages)
            for page in pages:
                if page not in liked_pages:
                    page_suggestion[page]=page_suggestion.get(page, 0)+ len(shared_pages)
    sorted_suggestions=sorted(page_suggestion.items(), key=lambda x:x[1], reverse=True)
    return [(page_id, score) for page_id, score in sorted_suggestions]
                      
load_data("messive_data.json")  
user_id=12
recc=page_recc(user_id, data)

for user in data["users"]:
    if user_id == user["id"]:
        print("reccomandetions for user:")
        print(user["name"], " can like :\n", recc)
    

user_id=12       
recommendations=rec(user_id, data)
name=[user["name"] for user in data["users"] if user_id == user["id"]]
print(f"{user_id}.{name} can follow: {recommendations}")
```
